```python
##importing dataset
import numpy as np
import pandas as pd
from matplotlib import pyplot as py
import warnings
warnings.filterwarnings('ignore')
```

```python
##loading dataset
data=pd.read_csv('Bitcoin_Price.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
data.drop(['Date'],1,inplace=True)
```

```python
data.head()
```

```python
##predicting 30 days in future
pred_days=30
data['Prediction']=data[['Price']].shift(-pred_days)
```

```python
data.head()
```

```python
data.tail()
```

```python
##seperate x and y
x=np.array(data.drop(['Prediction'],1))
x=x[: len(data) - pred_days]
```

```python
x.shape
```

```python
x
```

```python
##creating y as np array
y=np.array(data['Prediction'])
y=y[: -pred_days]
```

```python
y.shape
```

```python
y
```

```python
##split the data
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
pred_days_array=np.array(data.drop(['Prediction'],1))[-pred_days: ]
```

```python
print(pred_days_array)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RF=RandomForestRegressor(n_estimators=1000,random_state=1)
RF.fit(x_train,y_train)
print('Random Forest Accuracy: {:.2f}%'.format(RF.score(x_test,y_test)*100))
```

```python
##prediction
RF_pred=RF.predict(x_test)
print(RF_pred)
```

```python
print(y_test)
```

```python
##model prediction for 30 days
RF_pred_30=RF.predict(pred_days_array)
print(RF_pred_30)
```
